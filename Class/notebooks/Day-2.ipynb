```python


def cls():
    os.system(command="clear")

pomodoro: bool = True
count = 0
study_time = 1
study_sec: int = study_time*60
while pomodoro:
    cls()
    print(f"Remaining Time : {study_sec}")
    time.sleep(1)
    study_sec -= 1
    count += 1
    if study_sec < 1:
        pomodoro = False



    

```
